#### Q4.C

Null Hypothesis = "Gender has no effect on smoking habits"   
Alternate Hypothesis = "Gender has an effect on smoking habits"   


```python
Data = pd.crosstab(Insurance['sex'],Insurance['smoker'])
```

```python
chi, p_value, dof, expected =  stats.chi2_contingency(Data)
```

```python
p_value
```

As p value is less than 0.05  so we will accept alternative hypotesis that is Gender has an effect on smoking habits

#### Q4.D

```python
Null hypothesis = "No. of children has no effect on bmi"   
Alternante hypothesis = "No. of children has an effect on bmi" 
```

```python
Women_df = Insurance[Insurance['sex']=='female']
Women_df
zero_children = Women_df[Women_df.children == 0]['bmi']
one_children = Women_df[Women_df.children == 1]['bmi']
two_children = Women_df[Women_df.children == 2]['bmi']
```

```python
f_stat, p_value = stats.f_oneway(zero_children,one_children,two_children)
```

```python
p_value
```

#### As P value is more than 0.05 so we will accept null hypothesis that is(No. of children has no effect on bmi)
